## Exploratory Data Analyis

Importamos librerías necesarias para el EDA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

Importamos el dataset

```python
df = pd.read_csv('datasets/AccidentesAviones.csv')
```

```python
df.head()
```

Inicialmente vemos que dentro del CSV está la columna de índices, por lo que eliminaremos la columna Unnamed: 0

```python
df.drop(columns='Unnamed: 0', inplace=True)
```

```python
df.head()
```

Ahora, cambiaremos los valores '?' por valores nulos para facilitar el análisis

```python
df.replace('?', np.nan, inplace=True)
```

```python
df.head()
```

Ahora procedemos a ver la información del dataset

```python
df.info()
```

Podemos ver que la columna con más valores nulos es flight_no, por lo que no nos aporta mucho valor al proyecto, entonces podemos eliminarla

```python
df.drop(columns='flight_no', inplace=True)
```

Eliminamos los valores nulos, ya que no nos aportan valor en la elaboración de KPIs

```python
df.dropna(inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.head()
```

El dataset queda con 3021 registros

```python
df.info()
```

Vemos que en los campos numéricos el type es object y no int, por lo que corregimos este aspecto

```python
conversiones = {'all_aboard' : int, 'PASAJEROS A BORDO' : int, 'crew_aboard' : int, 
                'cantidad de fallecidos' : int, 'passenger_fatalities' : int, 'crew_fatalities' : int, 'ground' : int}

df = df.astype(conversiones)
```

```python
df.info()
```

Hay un problema con la columna HORA declarada y es que hay algunos registros que no están en un formato adecuado, por lo que vamos a ver cuáles registros no cumplen con el formato

```python
def es_entero(valor):
    try:
        int(valor)
        return True
    except ValueError:
        return False

# Filtrar los registros que no se pueden convertir a int
registros_no_enteros = df[~df['HORA declarada'].apply(es_entero)]

# Mostrar los registros que no se pueden convertir a int
registros_no_enteros
```

Procedemos a normalizar los registros defectuosos

```python
df['HORA declarada'] = df['HORA declarada'].str.replace(':', '')
df['HORA declarada'] = df['HORA declarada'].str.replace('c ', '')
```

Volvemos a aplicar la función

```python
def es_entero(valor):
    try:
        int(valor)
        return True
    except ValueError:
        return False

# Filtra los registros que no se pueden convertir a int
registros_no_enteros = df[~df['HORA declarada'].apply(es_entero)]

# Muestra los registros que no se pueden convertir a int
registros_no_enteros
```

Ya solo nos quedan 8 valores por normalizar

```python
df['HORA declarada'] = df['HORA declarada'].str.replace('Z', '')
df['HORA declarada'] = df['HORA declarada'].str.replace('c', '')
df['HORA declarada'] = df['HORA declarada'].str.replace(';', '')
```

Ahora sí, podemos convertir la columna HORA declarada a int

```python
df['HORA declarada'] = df['HORA declarada'].astype(int)
```

Aplicamos el formato para que todos los números sean de 4 dígitos

```python
def formatear_numero(numero):
    return f"{numero:04}"

df['HORA declarada'] = df['HORA declarada'].apply(formatear_numero)
```

```python
df.tail()
```

```python
df['fecha'] = pd.to_datetime(df['fecha'], format='%B %d, %Y')
```

```python
df.head()
```

Ahora, crearemos una columna que contenga en formato de lista las palabras de la columna Ruta utilizando un split

```python
df['ubicacion'] = df['Ruta'].str.split(', ')
```

```python
df.head()
```

Creamos una columna para poner el país

```python
df['pais'] = pd.Series([])
```

Generalmente el último elemento de la lista es el país, por lo tanto asignamos este valor a la columna país

```python
df['ubicacion']
```

```python
for i, lista in enumerate(df['ubicacion']):
    df['pais'][i] = df['ubicacion'][i][-1]
```

```python
df.head()
```

Nos encontramos con un problema y es que los accidentes que ocurrieron en EEUU no especifican el país, sino la ciudad y el estado, por lo que el último elemento de la lista corresponde al estado de EEUU donde ocurrió el accidente, por lo que vamos a arreglar esta parte

```python
us_states = [
    "Alabama", "Alaska", "Arizona", "Arkansas", "California",
    "Colorado", "Connecticut", "Delaware", "Florida", "Georgia",
    "Hawaii", "Idaho", "Illinois", "Indiana", "Iowa",
    "Kansas", "Kentucky", "Louisiana", "Maine", "Maryland",
    "Massachusetts", "Michigan", "Minnesota", "Mississippi", "Missouri",
    "Montana", "Nebraska", "Nevada", "New Hampshire", "New Jersey",
    "New Mexico", "New York", "North Carolina", "North Dakota", "Ohio",
    "Oklahoma", "Oregon", "Pennsylvania", "Rhode Island", "South Carolina",
    "South Dakota", "Tennessee", "Texas", "Utah", "Vermont",
    "Virginia", "Washington", "West Virginia", "Wisconsin", "Wyoming"
]

for i, pais in enumerate(df['pais']):
    if pais in us_states:
        df['pais'][i] = 'United States'
```

Finalmente, quitamos la columna ubicación ya que no es necesaria

```python
df.drop(columns='ubicacion', inplace=True)
```

Ahora que los datos están normalizados, ejecutamos la función describe para ver percentiles, media y desviación estándar

```python
df.describe()
```

Creamos boxplots para detectar outliers

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 4))

ax[0].boxplot(df['cantidad de fallecidos'])
ax[0].set_title('Cantidad de fallecidos')
ax[0].set_ylim(0, 200)

ax[1].boxplot(df['all_aboard'])
ax[1].set_title('Pasajeros')
ax[1].set_ylim(0, 400)

plt.show()
```

Según estos gráficos vemos que hay muchos valores atípicos en ambas columnas, pero al analizar la tabla anterior vemos que hay vuelos de hasta 614 pasajeros a bordo por lo que si bien estos datos pueden parecer outliers, no están del todo incorrectos. En este link (https://www.nlarenas.com/2021/08/record-maximo-pasajeros-llevado-un-avion/) podemos ver que se hablan de vuelos de hasta 1122 pasajeros por lo que si es común que hayan vuelos de gran cantidad de pasajeros, por lo tanto no se eliminarán estos datos.

Ahora vemos la cantidad de accidentes aéreos por década

```python
df['decada'] = df['fecha'].dt.year // 10 * 10

accidentes_por_decada = df.groupby('decada')['all_aboard'].count()

plt.figure(figsize=(10, 6))
accidentes_por_decada.plot(kind='bar', color='skyblue')
plt.title('Total de Accidentes Aéreos por Década')
plt.xlabel('Década')
plt.ylabel('Total de Accidentes')
plt.xticks(rotation=45)
plt.show()
```

En este gráfico podemos ver la cantidad de muertes por década, se puede evidenciar que la mayor cantidad de accidentes ocurrieron entre los años 1950 y 2000, vemos que después de estos años se presentó una disminución considerable, esto se lo podemos atribuir a los grandes avances tecnológicos en materia de tecnología y seguridad.

Hacemos un pariplot para analizar gráficamente las posibles relaciones entre las variables

```python
sns.pairplot(df)
```

A simple vista podemos ver relaciones directas entre la cantidad de pasajeros y la cantidad de fallecidos, lo cual es de esperarse. También podemos ver en cuanto a las distribuciones de las variables con respecto a las décadas que después de 1960 hubo más accidentes en aviones de gran cantidad de pasajeros, número que empieza a disiminuir a partir del año 2000.

Hacemos una matriz de correlación entre las variables numéricas para identificar relaciones inesperadas entre estas 

```python
columnas = ['fecha', 'HORA declarada', 'all_aboard', 'PASAJEROS A BORDO', 'crew_aboard', 'cantidad de fallecidos', 'passenger_fatalities', 'crew_fatalities', 'ground']

correlation_matrix = df[columnas].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True)

plt.title('Matriz de Correlación')
plt.xticks(rotation=45)

plt.show()
```

Podemos ver que no hay correlaciones inesperadas, pues es de esperarse que la cantidad de fallecidos tenga una alta correlación con la passenger_fatalities, crew_fatalities, pasajeros a bordo y all_aboard, pues a mayor cantidad de personas dentro del avión es de esperarse que haya un mayor número de fallecidos (como pudimos ver en el pairplot)
